## âœ… Step 1: Verify Installation

````python
# ğŸ§ª Installation Test - Run this cell first!

print("ğŸ” Checking TATty Agent installation...")

try:
    # Test import
    from tatty_agent.tests import test_basic_functionality
    print("âœ… TATty Agent found!")
    
    print("\nğŸ§ª Running basic functionality test...")
    test_basic_functionality()
    print("\nğŸ‰ Everything works! Ready to go!")
    
except ImportError as e:
    print("âŒ TATty Agent not found!")
    print("\nğŸ“¦ Please install it first:")
    print("   pip install TATty-agent[full]")
    print("\nThen restart this kernel and try again.")
    print(f"\nError details: {e}")
    
except Exception as e:
    print("âš ï¸  Installation found but there's an issue:")
    print(f"   {e}")
    print("\nğŸ’¡ This might still work - continue to next steps!"))
````

## ğŸ¤– Step 2: Your First Agent

````python
# ğŸ¤– Create Your First Agent

print("ğŸ”§ Creating TATty Agent...")

try:
    from tatty_agent import TattyAgent
    
    # Create your agent (verbose=False for cleaner output)
    agent = TattyAgent(verbose=False)
    
    print("âœ… Agent created successfully!")
    print(f"ğŸ“ Working directory: {agent.working_dir}")
    print(f"ğŸ’¬ Ready for your questions!")
    
except Exception as e:
    print(f"âŒ Failed to create agent: {e}")
    print("\nğŸ’¡ Make sure TATty Agent is properly installed:")
    print("   pip install TATty-agent[full]")
````

## ğŸ’¬ Step 3: Ask Your First Question

````python
# ğŸ’¬ Your First Conversation

print("ğŸ—£ï¸  Asking the agent a simple question...")

try:
    # Simple question that should work without complex dependencies
    result = agent.ask("Hello! What can you help me with in this notebook?")
    print("\nğŸ¤– Agent says:")
    print(result)
    
except NameError:
    print("âŒ Agent not created yet - please run the cell above first!")
    
except Exception as e:
    print(f"âš ï¸  Agent responded with an issue: {e}")
    print("\nğŸ’¡ This might be normal - the core functionality should still work!")
    print("Let's continue to magic commands...")
````

## âœ¨ Step 4: Magic Commands (The Fun Part!)

````python
# âœ¨ Load Magic Commands

print("ğŸª„ Loading TATty magic commands...")

try:
    # Load TATty magic commands
    %load_ext tatty_agent.jupyter.magic
    print("âœ… Magic commands loaded successfully!")
    print("Now you can use:")
    print("  â€¢ %tatty \"your question\"")
    print("  â€¢ %%tatty (for multi-line)")
    
except Exception as e:
    print(f"âŒ Magic commands failed to load: {e}")
    print("\nğŸ’¡ You can still use the agent directly:")
    print("   agent.run(\"your question\")")
    print("\nThis is often due to missing optional dependencies.")
    print("Try: pip install TATty-agent[jupyter]")
````

````python
# Use magic command for quick queries
%tatty "What files are in the current directory?"
````
